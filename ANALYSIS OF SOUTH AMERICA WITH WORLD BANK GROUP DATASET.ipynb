```python
for i in df_SA.columns:
    if df_SA[i].isnull().any():
        df_SA[i+"_missing"] = 0
        df_SA.loc[df_SA[i].isnull(),i+"_missing"]=1
        
df_SA    
```

```python
df_SA['Income_420']= df_SA['Income_420'].fillna(np.mean(df_SA['Income_420']))
df_SA['Income_H20']= df_SA['Income_H20'].fillna(np.mean(df_SA['Income_H20']))
df_SA['Income_L20']= df_SA['Income_L20'].fillna(np.mean(df_SA['Income_L20']))
df_SA['Income_220']= df_SA['Income_220'].fillna(np.mean(df_SA['Income_220']))
df_SA['Income_320']= df_SA['Income_320'].fillna(np.mean(df_SA['Income_320']))
df_SA['Industry']= df_SA['Industry'].fillna(np.mean(df_SA['Industry']))
df_SA['MilitaryExp']= df_SA['MilitaryExp'].fillna(np.mean(df_SA['MilitaryExp']))
df_SA['PopLarge']= df_SA['PopLarge'].fillna(np.mean(df_SA['PopLarge']))
df_SA['ExtremePoverty']= df_SA['ExtremePoverty'].fillna(np.mean(df_SA['ExtremePoverty']))
df_SA['Poverty']= df_SA['Poverty'].fillna(np.mean(df_SA['Poverty']))
df_SA['NationalPoverty']= df_SA['NationalPoverty'].fillna(np.mean(df_SA['NationalPoverty']))
df_SA['HIV']= df_SA['HIV'].fillna(np.mean(df_SA['HIV']))
df_SA['Services_ValAdd']= df_SA['Services_ValAdd'].fillna(np.mean(df_SA['Services_ValAdd']))
df_SA['Undernourishment']= df_SA['Undernourishment'].fillna(np.mean(df_SA['Undernourishment']))
df_SA['TaxRev']= df_SA['TaxRev'].fillna(np.mean(df_SA['TaxRev']))
df_SA['Underweight_5y']= df_SA['Underweight_5y'].fillna(np.mean(df_SA['Underweight_5y']))
df_SA['LitRate_y']= df_SA['LitRate_y'].fillna(np.mean(df_SA['LitRate_y']))
df_SA['LitRate_a']= df_SA['LitRate_a'].fillna(np.mean(df_SA['LitRate_a']))
df_SA['GovExp_Edu']= df_SA['GovExp_Edu'].fillna(np.mean(df_SA['GovExp_Edu']))
```

```python
df_MV = pd.read_excel("missing values.xlsx",index_col=0)

df_SA = df_SA.fillna(df_MV)

print(df_SA.isnull().any())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd


file = 'WDIW South America.xlsx'
SouthAm = pd.read_excel(file)

# See general values 
print(SouthAm.describe().round(2))

#Creating Heatmap to observe Correlations
fig, ax = plt.subplots(figsize=(30,30))

sns.heatmap(data = SouthAm.corr(),
            cmap = 'coolwarm',
            square = True,
            annot = False,
            linecolor = 'black',
            linewidths = 0)
plt.tight_layout()
plt.savefig('HeatMap.png')
plt.show()
```
